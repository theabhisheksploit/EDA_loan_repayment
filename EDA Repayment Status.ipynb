# EDA Repayment Status

### This a Exploratory Data Analysis for Repayment status Based On Different characteristics of user.

#### EDA for loan or credit-related datasets, "repayment status" typically refers to the historical record of how borrowers have repaid their loans over time. This information is crucial for lenders and financial institutions to assess a borrower's creditworthiness and to understand their ability to manage debt responsibly.

## Step 1 : Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
# pd.set_options('max_columns', 200)
```

## Step 2 : Load the dataset

```python
file_path = 'Data_Analyst_Task.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Display the first few rows of the DataFrame to check if it loaded successfully
df.head()
```

#### Note : In the above data types we saw some wrong data type like  DOB, Borrow Date etc. so we need to change these data types.

```python
''' 
lets move forward to improve the data type of each columns in the dataframe, we can do this in Data Cleaning and Preprocessing section 
but lets do it here.
'''
# Improving "DOB" Data Type:
df['DOB'] = pd.to_datetime(df['DOB'])

# Improving "Borrow Date" Data Type:
df['Borrow date'] = pd.to_datetime(df['Borrow date'])

# Improving "Employment Type" Data Type:
df['Employment Type'] = df['Employment Type'].astype('category')

# Improving "Gender" Data Type:
df['gender'] = df['gender'].astype('category')

# Improving "Purpose of the Loan" Data Type:
df['Purpose of the loan'] = df['Purpose of the loan'].astype('category')

# Improving "Salary" Data Type we get an error that "salary" column have a invalid character which can't we convert into float "-"
df['salary'] = df['salary'].replace([ '-'], '', regex=True)
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
print(df["salary"].dtype)
```

```python
# lets check again the data types of each columns to know more about data
df.dtypes
```

```python
# Now we can check the overview of some statistical analysis 
df.describe()
```

```python
# Check the data types and missing values in the dataset
df.info()
```

## Step 4 : Data Cleaning and Preprocessing
               * Dropping irrelevant column and rows
               * Identifying duplicated columns
               * checking null value
               * rename the columns name

```python
# Drop the default index column
df = df.drop(columns='Index')
```

```python
df.info()
```

```python
# check the Duplicate value
df.duplicated()
```

```python
# check the missing values once again
df.isna().sum()
```

#### In the above we can see there is 233 missing values

```python
# Calculate the mean of the non-missing values
mean_salary = df["salary"].dropna().mean()

# Impute the missing values with the mean
df["salary"].fillna(mean_salary, inplace=True)
```

```python
# check the once missing values once again
df.isna().sum()
```

```python
# Rename the columns name 
df = df.rename(columns={'Purpose of the loan' : 'Loan Purpose'})
```

## Step : EDA for repayment status based on different characteristics
              

```python
# Employement Sectors repayment status
# Plot the data
sns.catplot(x="Employement Sector", hue="Repayment Status", data=df, kind="count")

# Add a title and rotate the x-ticks
plt.title("Repayment Status by Employment Sectors")
plt.xticks(rotation=90)

plt.show()
```

```python
# Loan Purpose with Repayment status
plt.figure(figsize=(10, 6))
sns.countplot(x='Loan Purpose', hue='Repayment Status', data=df)
plt.title('Repayment Status by Loan Purpose')
plt.xticks(rotation=45)
plt.show()
```

```python
# State
plt.figure(figsize=(12, 6))
sns.countplot(x='State', hue='Repayment Status', data=df)
plt.title('Repayment Status by State')
plt.xticks(rotation=90)
plt.show()

```

```python
top_10_cities = df['City'].value_counts().head(10)
print(top_10_cities)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_10_cities, labels=top_10_cities.index, autopct='%1.1f%%')
plt.title('Top 10 Cities')
plt.show()
```

```python
# Salary 
plt.figure(figsize=(10, 6))
sns.boxplot(x='Repayment Status', y='salary', data=df)
plt.title('Repayment Status by salary')
plt.show()
```

```python
# Extract month and quarter from the 'Date' column for repayment status by month and quarter
df['Month'] = df['Borrow date'].dt.month
df['Quarter'] = df['Borrow date'].dt.quarter
df.info()
```

```python
# Month with repayment status
plt.figure(figsize=(10, 6))
sns.countplot(x='Month', hue='Repayment Status', data=df)
plt.title('Repayment Status by Month')
plt.xticks(rotation=45)
plt.show()

# quarter 

plt.figure(figsize=(8, 5))
sns.countplot(x='Quarter', hue='Repayment Status', data=df)
plt.title('Repayment Status by Quarter')
plt.show()
```

```python
# Age and Gender with Repayment Status
''' for that we need to calculate age'''

from datetime import datetime

# Calculate current date
current_date = datetime.now()

# Calculate age by subtracting the DOB from the current date and extracting the years
df['Age'] = (current_date - df['DOB']).astype('<m8[Y]')

```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='gender', kde=True, multiple='stack')
plt.title('Distribution of Age by Gender')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x='gender', hue='Repayment Status', data=df)
plt.title('Repayment Status by Gender')
plt.show()
```

```python
# Approval Category with Repayment Status
plt.figure(figsize=(8, 5))
sns.countplot(x='Approval Category', hue='Repayment Status', data=df)
plt.title('Repayment Status by Approval Category')
plt.show()
```

# Additional insights

```python
# The most common Employment Sectors and their corresponding repayment pattern.
# Group the data by 'Employment Sector' and 'Repayment Status' and count the occurrences
sector_repayment_counts = df.groupby(['Employement Sector', 'Repayment Status']).size().reset_index(name='Count')

# Normalize the counts to get percentages
sector_repayment_counts['Percentage'] = sector_repayment_counts.groupby('Employement Sector')['Count'].apply(lambda x: x / x.sum() * 100)

# Create a bar plot to visualize the most common Employment Sectors and their repayment patterns
plt.figure(figsize=(10, 6))
sns.barplot(x='Employement Sector', y='Percentage', hue='Repayment Status', data=sector_repayment_counts)
plt.title('Repayment Status by Employment Sectors')
plt.xlabel('Employement Sector')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.legend(title='Repayment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

###  Multivariate Analysis
   Explore relationships involving more than two variables.

```python
# Set a custom color palette for better distinction in scatterplot
scatter_palette = sns.color_palette('Set1', n_colors=3)

# Scatterplot (Age vs. User's Salary by Repayment Status)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y="salary", hue='Repayment Status', palette=scatter_palette, data=df)
plt.title('Age vs Salary by Repayment Status', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('User\'s Salary', fontsize=12)

# Move the legend outside the plot area to the upper left
plt.legend(title='Repayment Status', fontsize=10, loc='upper left', bbox_to_anchor=(1.01, 1))

plt.tight_layout()
plt.show()

# Set a custom color palette for better distinction in countplot
count_palette = sns.color_palette('pastel')

# Countplot (Categorical vs. Categorical vs. Categorical - Gender vs. Employment Sectors vs. Repayment Status)
g = sns.catplot(x='gender', hue='Employement Sector', col='Repayment Status', kind='count', palette=count_palette, data=df)
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle('Repayment Status by Gender and Employment Sectors', fontsize=16)
g.set_axis_labels('gender', 'Count')
g.set_titles('Repayment Status: {col_name}')
plt.tight_layout()
plt.show()

```
